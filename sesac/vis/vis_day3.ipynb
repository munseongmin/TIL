```python
import seaborn as sns #데이터 제공, 열 지도(히트 맵)등 시각화
iris = sns.load_dataset("iris")
```

```python
#붓꽃 종별 특징별로 최대값 - 최소값으로 데이터프레임 구성
```

```python
def myfunc(x):
    return x.max() - x.min()
```

```python
grs.agg(myfunc) #agg() : grs의 그룹별로 괄호안에 myfunc 함수 적용
```

```python
iris.groupby(iris.species).describe()
```

```python
#iris에 대해 종별로 구분한 다음 petal_length값 기준으로 내림차순
```

```python
iris.groupby(iris.species).get_group('setosa').sort_values(by='petal_length', ascending=False)
```

```python
def myfunc2(df):
    return df.sort_values(by='petal_length', ascending=False)

iris.groupby(iris.species).apply(myfunc2) #데이터프레임에서 그룹단위로 함수 적용하려고 할때 apply 함수 사용
```

```python
def q3cut(s):
    return pd.qcut(s, 3, labels=["소", "중", "대"]).astype(str)


iris["petal_length_class"] = iris.groupby(iris.species).petal_length.transform(q3cut) #transform() : iris 데이터프레임이 변함
iris[["petal_length", "petal_length_class"]].tail(10)
```

```python
data = {
    "도시": ["서울", "서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [1, 9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]
df1 = pd.DataFrame(data, columns=columns)
df1
```

```python
#pivot_table(데이터 행 열)
df1.pivot_table('인구', '도시', '연도', aggfunc='mean') #aggfunc='mean'평균값이라는 함수가 디폴트
```

```python
df1.pivot_table('인구', '도시', '연도', aggfunc='max')
df1.pivot_table('인구', '도시', '연도', aggfunc='min')
df1.pivot_table('인구', '도시', '연도', aggfunc='std')
df1.pivot_table('인구', '도시', '연도', aggfunc='var')
df1.pivot_table('인구', '도시', '연도', aggfunc='count')
df1.pivot_table('인구', '도시', '연도', aggfunc='sum')
```

```python
def add_prop(g):
    g['prop']=g['births'] / g['births'].sum()
    return g
```

```python
names=names.groupby(['year','sex']).apply(add_prop)
```

```python
names
```

```python
names[(names['year']==1880) & (names['sex']=='F')]['prop'].sum()
```

```python
def get_top1000(g):
    return g.sort_values('births', ascending=False)[:1000]
```

```python
top1000=names.groupby(['year','sex']).apply(get_top1000)
top1000
```
