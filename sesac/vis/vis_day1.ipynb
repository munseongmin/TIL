```python
'''
변수 : 연속변수, 범주변수(이산변수)
척도(scale) : 변수값을 표현하는 수준

1) 명명척도(nominal scale) : 종류, 순서x
    ex)혈액형

2) 서열척도(ordinal scale) : 종류, 순서o, 종류들의 간격이 동일x
    ex)직급, 학점, 메달 등
    
3) 등간척도 : 수치형 자료, 간격 일정
    ex)섭씨온도 - 영상 10도가 영상 2도 보다 수치적으로 5배 크지만 실제로 5배 따뜻하지는 않다.
    
4) 비율척도 : 등간척도 + 절대영점(영점의 의미가 존재하지 않는 상태)
    ex)길이 - 10미터는 2미터보다 수치적으로 5배 크고, 실제로도 5배 길다.
    
'''
```

```python
#데이터프레임 합치기 함수
```

```python
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7), dtype="Int64")})
df2 = pd.DataFrame({"key": ["a", "b", "d"],
                    "data2": pd.Series(range(3), dtype="Int64")})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2) #두 데이터프레임에 관해서 이름이 같은 것끼리 합치기, 이름이 같은 열이 없으면 에러
```

```python
pd.merge(df1,df2, on='key') #merge 함수에 on이라는 옵션이 디폴트값
```

```python
df3 = pd.DataFrame({"lkey": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7), dtype="Int64")})
df4 = pd.DataFrame({"rkey": ["a", "b", "d"],
                    "data2": pd.Series(range(3), dtype="Int64")})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3,df4, left_on='lkey', right_on='rkey') #같은 이름의 열인덱스 없어도 합치기
```

```python
pd.merge(df3,df4, left_on='lkey', right_on='rkey', how='outer') #모든 데이터 합치기 결측값 문자 NaN, 숫자 <NA>
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how='outer')
```
