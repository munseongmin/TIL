```python
#vote average가 vote count에 따라서 보정해서 score에 저장
def weighted_rating(x, m=m, C=C): #weighted_rating(x, 568, 7.03)
    v = x['vote_count']
    R = x['vote_average']
    
    return ( v / (v+m) * R ) + (m / (m + v) * C)
```

```python
data['score'] = data.apply(weighted_rating, axis = 1)
```
