```python
#chipo 에서 order_id가 같은 데이터 끼리 묶어서 트랜잭션으로 구성
chipo=pd.read_csv("chipotle.csv") #tsv
chipo
```

```python
df=chipo[['order_id', 'item_name']]
df
```

```python
df['order_id'].max()
```

```python
df_arr=[[]for i in range(1, df['order_id'].max()+1)] #1~1834+1
#[[], [], [], ................., []] => TransactionEncoder
```

```python
np.shape(df_arr)
```

```python
#df_arr에 i가 append 되어야 함
n=0
for i in df['item_name']:
    df_arr[df['order_id'][n]-1].append(i)
    n+=1
```

```python
df_arr
```

```python
max(len(lst) for lst in df_arr)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
te=TransactionEncoder()
te_arr=te.fit(df_arr).transform(df_arr)
```

```python
te_arr
te_df=pd.DataFrame(te_arr, columns=te.columns_)
```

```python
te_df
```

```python
freq_items=apriori(te_df, min_support=0.02, use_colnames=True)
freq_items
```

```python
ar_df=association_rules(freq_items, metric='lift', min_threshold=1) #최소임계값 1 => lift가 1 이상인 것만 출력
ar_df
```

```python
ar_df.info()
```

```python
#rules=apriori(transactions=df_time['items'].tolist(), min_support=0.0003, min_confidence=0.01)
rules=apriori(transactions=df_time['items'].tolist(), min_support=0.002, min_confidence=0.02, min_lift=5, min_length=2, max_length=2)
```

```python
res=list(rules)
```

```python
#선구매 후구매 지지도 신뢰도 향상도
```

```python
tuple(res[0][2][0][0])[0]
tuple(res[0][2][0][1])[0]
```

```python
pre=[tuple(r[2][0][0])[0] for r in res]
con=[tuple(r[2][0][1])[0] for r in res]
```

```python
#support res[0][1]
supports=[r[1] for r in res]
#confidence res[0][2][1][2]
confidences=[r[2][1][2] for r in res]
#lift
lifts=[r[2][1][3] for r in res]
res_df=pd.DataFrame(list(zip(pre, con, supports, confidences, lifts)), columns=['pre', 'con', 'Support', 'Confidence', 'Lift'])
res_df
```
