```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model=KNeighborsRegressor(n_neighbors=5, weights='distance')
model
```

```python
model.fit(movie_dataset, movie_ratings) #가지고 있는 데이터셋, 예측하고자 하는 데이터
```

```python
#1892, 2 => ratings?
model.predict([[1892,2]])
```

```python
#2020, 6 => ratings?, 1900, 1 => ratings?
model.predict([[2020,6],[1900,1]])
```
